```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
cities = ('Bogor', 'Bandung', 'Jakarta', 'Semarang', 'Yogyakarta', 
          'Surakarta','Surabaya', 'Medan', 'Makassar')
 
populations = (45076704, 11626410, 212162757, 19109629, 50819826, 17579085,
               3481, 287750, 785409)
```

```python
plt.bar(x=cities, height=populations)
plt.show()
```

```python
plt.bar(x=cities, height=populations)
plt.xticks(rotation=45)
plt.show()
```

```python
plt.barh(y=cities, width=populations)
plt.show()
```

```python
df = pd.DataFrame({
    'Cities' : cities,
    'Population' : populations,
})
```

```python
df.sort_values(by='Population', inplace=True)
```

```python
plt.barh(y=df["Cities"], width=df["Population"])
plt.show()
```

```python
plt.barh(y=df["Cities"], width=df["Population"])
plt.xlabel("Population (Millions)")
plt.title("Population of Cities in Indonesia")
plt.show()
```

```python
sns.barplot(y=df["Cities"], x=df["Population"], orient="h",color="yellow")
plt.xlabel("Population (Millions)")
plt.title("Population of Cities in Indonesia")
plt.show()
```

#### Donut and Pie Plot

```python
flavors = ('Chocolate', 'Vanilla', 'Macha', 'Others')
votes = (50, 20, 30, 10)
colors = ('#8B4513', '#FFF8DC', '#93C572', '#E67F0D')
explode = (0.05, 0.01, 0.01, 0.01)
```

```python
plt.pie(
    x=votes,
    labels=flavors,
    autopct='%1.1f%%',
    colors=colors,
    explode=explode
)
plt.show()
```

```python
plt.pie(
    x=votes,
    labels=flavors,
    colors=colors,
    wedgeprops= {'width': 0.4}
)
```
